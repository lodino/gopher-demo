# Preparation

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# ignore all the warnings
import warnings
warnings.filterwarnings('ignore') 
```

```python
f = lambda x: np.log(np.cosh(x))
f_prime = lambda x: 2/(1+np.exp(-2*x))-1
f_prime_prime = lambda x: 4*np.exp(-2*x)/(1+np.exp(-2*x))**2
Q = lambda x, y: f(y) + f_prime(y)*(x-y) + 1/2*(x-y)**2
```

# (c)

```python
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x), label='f(x)')
plt.plot(x, Q(x, -2), label='Q(x, -2)')
plt.plot(x, Q(x, 3), label='Q(x, 3)')
plt.legend()
plt.show()
```

# (f)

$$x_{n+1}=x_n-f'(x_n)$$

## Start from x=-2

```python
plt.figure(figsize=(10, 4))
plt.subplot(121)
num_iter = 5
x_n = -2 # initial state, n=0
x_n_ls = [x_n]
x = np.arange(-4, 4, 0.1)
plt.plot(x, f(x), label='f(x)')
for i in range(num_iter):
    x_n_plus_1 = x_n - f_prime(x_n)
    x_n_ls.append(x_n_plus_1)
    plt.plot([x_n, x_n_plus_1], [f(x_n), f(x_n_plus_1)], 'r:x')
    x_n = x_n_plus_1
plt.legend()
plt.title('Start from x=-2')
plt.subplot(122)
plt.plot(np.arange(0, num_iter+1), x_n_ls)
plt.xlabel('n')
plt.ylabel('$x_n$')
plt.tight_layout()
plt.show()
for n, x_n in enumerate(x_n_ls):
    print(f'iter {n}: x_n={round(x_n, 4)}')
```

# (g)

$$x_{n+1}=x_n-\frac{f'(x_n)}{f''(x_n)}=x_n-\frac{1-e^{-4x_n}}{4e^{-2x_n}}\\|x_0|>|x_1|\Rightarrow |x_0|<1.08866$$

## Start from x=-2

```python
plt.figure(figsize=(10, 4))
plt.subplot(121)
num_iter = 5
x_n = -2 # initial state, n=0
x_n_ls = [x_n]
x = np.arange(-4, 4, 0.1)
plt.plot(x, f(x), label='f(x)')
for i in range(num_iter):
    x_n_plus_1 = x_n - f_prime(x_n)/f_prime_prime(x_n)
    x_n_ls.append(x_n_plus_1)
    plt.plot([x_n, x_n_plus_1], [f(x_n), f(x_n_plus_1)], 'r:x')
    x_n = x_n_plus_1
plt.legend()
plt.title('Start from x=-2')
plt.subplot(122)
plt.plot(np.arange(0, num_iter+1), x_n_ls)
plt.xlabel('n')
plt.ylabel('$x_n$')
plt.tight_layout()
plt.show()
for n, x_n in enumerate(x_n_ls):
    print(f'iter {n}: x_n={round(x_n, 4)}')
```

Does not converge, because $|x_0|>1.08866$
